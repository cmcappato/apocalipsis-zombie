Imports

```python
import pandas as pd
```

Leer archivo

```python
df = pd.read_csv("datasets/covid_data.csv")
df.head()
```

Modificación de variables

```python
# En vez de fallecido o no, cambio la variable a 'zombificado'. Esta será nuestra variable objetivo
df['sobreviviente'] = df['DATE_DIED'].apply(lambda x: 0 if x != '9999-99-99' else 1)
```

```python
df_zombie = df.rename(columns={
    'USMER': 'zona_infectada',
    'MEDICAL_UNIT': 'unidad_refugio',
    'SEX': 'genero',
    'PATIENT_TYPE': 'tipo_contacto',
    'DATE_DIED': 'fecha_zombificacion',
    'INTUBED': 'mordido',
    'PNEUMONIA': 'infeccion_zombie',
    'AGE': 'edad',
    'PREGNANT': 'embarazada',
    'DIABETES': 'condicion_base_diabetes',
    'COPD': 'condicion_base_epoc',
    'ASTHMA': 'condicion_base_asma',
    'INMSUPR': 'inmunidad_baja',
    'HIPERTENSION': 'condicion_base_hipertension',
    'OTHER_DISEASE': 'otras_condiciones',
    'CARDIOVASCULAR': 'condicion_base_cardio',
    'OBESITY': 'movilidad_reducida',
    'RENAL_CHRONIC': 'condicion_base_renal',
    'TOBACCO': 'resistencia_baja',
    'CLASIFFICATION_FINAL': 'confirmacion_infeccion',
    'ICU': 'unidad_critica'
})

df_zombie.head()
```

| Columna original       | Nombre zombie propuesto       | Interpretación en el contexto de apocalipsis zombie                      |
| ---------------------- | ----------------------------- | ------------------------------------------------------------------------ |
| `USMER`                | `zona_infectada`              | 1 = zona altamente expuesta al virus zombie                              |
| `MEDICAL_UNIT`         | `unidad_refugio`              | Tipo de instalación médica/refugio                                       |
| `SEX`                  | `genero`                      | Igual.                                                                   |
| `PATIENT_TYPE`         | `tipo_contacto`               | 1 = contacto directo (hospitalizado), 2 = contacto leve (ambulatorio)    |
| `DATE_DIED`            | `fecha_zombificacion`         | Fecha de zombificación (si falleció), si es `9999-99-99` → sobreviviente |
| `INTUBED`              | `mordido`                     | 1 = fue mordido, 2 = no                                                  |
| `PNEUMONIA`            | `infeccion_zombie`            | 1 = sí, 2 = no                                                           |
| `AGE`                  | `edad`                        | Igual.                                                                   |
| `PREGNANT`             | `embarazada`                  | 1 = sí, 2 = no, 97 = no aplica                                           |
| `DIABETES`             | `condicion_base_diabetes`     | 1 = sí, 2 = no                                                           |
| `COPD`                 | `condicion_base_epoc`         | 1 = sí, 2 = no                                                           |
| `ASTHMA`               | `condicion_base_asma`         | 1 = sí, 2 = no                                                           |
| `INMSUPR`              | `inmunidad_baja`              | 1 = sí, 2 = no                                                           |
| `HIPERTENSION`         | `condicion_base_hipertension` | 1 = sí, 2 = no                                                           |
| `OTHER_DISEASE`        | `otras_condiciones`           | 1 = sí, 2 = no                                                           |
| `CARDIOVASCULAR`       | `condicion_base_cardio`       | 1 = sí, 2 = no                                                           |
| `OBESITY`              | `movilidad_reducida`          | 1 = sí, 2 = no                                                           |
| `RENAL_CHRONIC`        | `condicion_base_renal`        | 1 = sí, 2 = no                                                           |
| `TOBACCO`              | `resistencia_baja`            | 1 = sí, 2 = no                                                           |
| `CLASIFFICATION_FINAL` | `confirmacion_infeccion`      | 1-3 = distintos niveles de confirmación del virus zombie                 |
| `ICU`                  | `unidad_critica`              | 1 = estuvo en UCI (caso grave), 2 = no                                   |


Guardo archivo final

```python
df_zombie.to_csv(path_or_buf="datasets/zombie_data.csv")
```
